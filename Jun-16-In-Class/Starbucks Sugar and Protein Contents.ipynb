```python
import csv 
import matplotlib.pyplot as plt # visulatization library
import numpy as np # math function

with open("starbucks_drinkMenu_expanded.csv") as csvfile:
    drinks_reader = csv.reader(csvfile)
    headers = next(drinks_reader, None) # skip over the headers
    
    # get the index that corresponds to the information we are interested in
    drink_category_index = headers.index("Beverage")
    sugars_index = headers.index(" Sugars (g)")
    
    # Store the sugar contents of various drinks
    sugar_in_lattes = []
    sugar_in_teas = []
    sugar_in_mochas = []
    
    for row in drinks_reader:
        drink_category = row[drink_category_index]
        sugar_grams = row[sugars_index]
        if "Latte" in drink_category:
            sugar_in_lattes.append(float(sugar_grams))
        if "Tea" in drink_category:
            sugar_in_teas.append(float(sugar_grams))
        if "Mocha" in drink_category:
            sugar_in_mochas.append(float(sugar_grams))
        
    beverage_categories = ["Latte", "Tea", "Mocha"]
    # average sugar content
    average_sugar_in_lattes = np.mean(sugar_in_lattes)
    average_sugar_in_teas = np.mean(sugar_in_teas)
    average_sugar_in_mochas = np.mean(sugar_in_mochas)

    average_sugars = [average_sugar_in_lattes, average_sugar_in_teas, average_sugar_in_mochas]

    vertical_bar_chart_figure = plt.figure() # the outter container
    vertical_bar_char_axes = vertical_bar_chart_figure.add_axes([0.1, 0.2, 0.8, 0.9]) # the actual chart inside the figure

    # create the bar chart using the bar() method
    # to specify x, y, and the colors for each bar
    vertical_bar_char_axes.bar(beverage_categories, average_sugars, color=["pink", "blue", "green"])
    
    
    # Customize the charts
    
    # Give a title 
    vertical_bar_char_axes.set_title("Vertical bar chart of average sugar in grams for different types of beverages at Starbucks")
    
    # Label axes
    vertical_bar_char_axes.set_ylabel("Average Sugar Content (g) grams")
    vertical_bar_char_axes.set_xlabel("Beverage Type")
    
    # Creating a horizontal bar graph
    horizontal_bar_chart_figure = plt.figure()
    horizontal_bar_chart_axes = horizontal_bar_chart_figure.add_axes([0.1, 0.2, 0.8, 0.9])
    
    # Creating the bar chart using barh method
    horizontal_bar_chart_axes.barh(beverage_categories, average_sugars, color=["pink", "blue", "green"])
    
    # Add title and axes labels
    horizontal_bar_chart_axes.set_title("Horizontal bar char of average sugar in grams for different types of beverages at Starbucks")
    horizontal_bar_chart_axes.set_xlabel("Average Sugar Content (g) grams")
    horizontal_bar_chart_axes.set_ylabel("Beverage Type")
    
    
    
```

```python
line_graph_figure = plt.figure()
line_graph_axes = line_graph_figure.add_axes([0.1, 0.2, 0.8, 0.9])

#create line graph using plot method
line_graph_axes.plot(beverage_categories, average_sugars)

# set title and axes labels
line_graph_axes.set_title("Line graph of average sugar in grams of different types of beverages at Starbucks")
line_graph_axes.set_ylabel("Average Sugar Content (g) grams")
line_graph_axes.set_xlabel("Beverage Type")
```

```python
import csv 
import matplotlib.pyplot as plt # visulatization library
import numpy as np # math function

with open("starbucks_drinkMenu_expanded.csv") as csvfile:
    drinks_reader = csv.reader(csvfile)
    headers = next(drinks_reader, None) # skip over the headers
    
    # get the index that corresponds to the information we are interested in
    drink_category_index = headers.index("Beverage")
    protein_index = headers.index(" Protein (g) ")
    
    # Store the sugar contents of various drinks
    protein_in_cappuccino = []
    protein_in_macchiato = []
    protein_in_smoothies = []
    
    for row in drinks_reader:
        drink_category = row[drink_category_index]
        protein_grams = row[protein_index]
        if "Cappuccino" in drink_category:
            protein_in_cappuccino.append(float(protein_grams))
        if "Macchiato" in drink_category:
            protein_in_macchiato.append(float(protein_grams))
        if "Smoothie" in drink_category:
            protein_in_smoothies.append(float(protein_grams))
        
    beverage_categories = ["Cappuccino", "Macchiato", "Smoothie"]
    # average sugar content
    average_protein_in_lattes = np.mean(protein_in_cappuccino)
    average_protein_in_teas = np.mean(protein_in_macchiato)
    average_protein_in_mochas = np.mean(protein_in_smoothies)

    average_protein = [average_protein_in_lattes, average_protein_in_teas, average_protein_in_mochas]

    vertical_bar_chart_figure = plt.figure() # the outter container
    vertical_bar_char_axes = vertical_bar_chart_figure.add_axes([0.1, 0.2, 0.8, 0.9]) # the actual chart inside the figure

    # create the bar chart using the bar() method
    # to specify x, y, and the colors for each bar
    vertical_bar_char_axes.bar(beverage_categories, average_protein, color=["red", "yellow", "orange"])
    
    
    # Customize the charts
    
    # Give a title 
    vertical_bar_char_axes.set_title("Vertical bar chart of average sugar in grams for different types of beverages at Starbucks")
    
    # Label axes
    vertical_bar_char_axes.set_ylabel("Average Protein Content (g) grams")
    vertical_bar_char_axes.set_xlabel("Beverage Type")
    
    # Creating a horizontal bar graph
    horizontal_bar_chart_figure = plt.figure()
    horizontal_bar_chart_axes = horizontal_bar_chart_figure.add_axes([0.1, 0.2, 0.8, 0.9])
    
    # Creating the bar chart using barh method
    horizontal_bar_chart_axes.barh(beverage_categories, average_protein, color=["red", "yellow", "orange"])
    
    # Add title and axes labels
    horizontal_bar_chart_axes.set_title("Horizontal bar char of average sugar in grams for different types of beverages at Starbucks")
    horizontal_bar_chart_axes.set_xlabel("Average Protein Content (g) grams")
    horizontal_bar_chart_axes.set_ylabel("Beverage Type")
    
    
line_graph_figure = plt.figure()
line_graph_axes = line_graph_figure.add_axes([0.1, 0.2, 0.8, 0.9])

#create line graph using plot method
line_graph_axes.plot(beverage_categories, average_protein)

# set title and axes labels
line_graph_axes.set_title("Line graph of average protein in grams of different types of beverages at Starbucks")
line_graph_axes.set_ylabel("Average Protein Content (g) grams")
line_graph_axes.set_xlabel("Beverage Type")
    
    
```

```python
# Creating histograms of the sugar content of various beverage types
number_of_bins = 10 # specify the number of bins/bars for the charts

latte_histogram_figure = plt.figure()
latte_histogram_axes = latte_histogram_figure.add_axes([0.1, 0.2, 0.8, 0.9])

# creating the histogram using hist method
latte_histogram_axes.hist(sugar_in_lattes, bins=number_of_bins)

# createing proper labels for x, y and title
latte_histogram_axes.set_title("Histogram of sugar content in Lattes")
latte_histogram_axes.set_xlabel("Sugar in grams")
latte_histogram_axes.set_ylabel("Frequency")

tea_histogram_figure = plt.figure()
tea_histogram_axes = tea_histogram_figure.add_axes([0.1, 0.2, 0.8, 0.9])

tea_histogram_axes.hist(sugar_in_teas, bins=number_of_bins)

tea_histogram_axes.set_title('Histogram of sugar content in Teas')
tea_histogram_axes.set_ylabel('Frequency')
tea_histogram_axes.set_xlabel('Sugar in grams')

mocha_histogram_figure = plt.figure()
mocha_histogram_axes = mocha_histogram_figure.add_axes([0.1, 0.2, 0.8, 0.9])

mocha_histogram_axes.hist(sugar_in_mochas, bins=number_of_bins)

mocha_histogram_axes.set_title('Histogram of sugar content in Mochas')
mocha_histogram_axes.set_ylabel('Frequency')
mocha_histogram_axes.set_xlabel('Sugar in grams')
```
