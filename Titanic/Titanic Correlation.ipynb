```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sn

df = pd.read_csv('titanic.csv')
df = df.drop(['PassengerId'], axis=1)

# Change male and femle into integer for comparision
genders = {"male": 1, "female": 0}
df['Sex'] = df['Sex'].map(genders)
sn.heatmap(df.corr(), annot=True)
plt.show()

# We can rapidly see the the strongest positive (Orange) and strongest negative correlations (Black). What are they?
# They are Pclass to Survied and Parch to SibSp
```
